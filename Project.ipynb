# Split data : X hasn't the target values, Y only contains target values

```python
X = df.iloc[:, 1:].values
Y = df.iloc[:, 0].values
```

# Normalize X data for KNN

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
X = mms.fit_transform(X)
```

# - With K-Fold cross validation

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from numpy import mean
from numpy import std
```

# Import models

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

# Stratified KFold initialisation (k=10 with shuffle)

```python
cv = StratifiedKFold(n_splits=10, shuffle = True, random_state=0)
```

# Random Forest Cross Validation and Mean Accuracy with confidence interval

```python
random_forest_scores = []

for train, test in cv.split(X,Y):
    
    random_forest = RandomForestClassifier(n_estimators = 100, criterion = 'entropy', random_state = 0)
    
    random_forest.fit(X[train], Y[train])
    random_forest_scores.append(random_forest.score(X[test], Y[test]))

print('Random Forest mean accuracy: %.3f (%.3f)' % (mean(random_forest_scores), std(random_forest_scores)))
```

# KNN Cross Validation and Mean Accuracy with confidence interval

```python
knn_scores = []

for train, test in cv.split(X,Y):
    
    knn = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2, weights = 'distance')
    
    knn.fit(X[train], Y[train])
    knn_scores.append(knn.score(X[test], Y[test]))

print('KNN mean accuracy: %.3f (%.3f)' % (mean(knn_scores), std(knn_scores)))
```

# - Without K-Fold cross validation

# Split data in train / test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
```

# KNN and Random Forest models initialisation and fit

```python
random_forest = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
random_forest.fit(X_train, Y_train)

knn = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2, weights='distance')
knn.fit(X_train, Y_train)
```

# KNN accuracy on test data and ROC AUC

```python
knn_prediction = knn.predict_proba(X_test)
knn_prediction = knn_prediction[:, 1]
knn_auc = roc_auc_score(Y_test, knn_prediction)
print('KNN : ROC AUC=%.3f' % (knn_auc))
knn_fpr, knn_tpr, __noNeed = roc_curve(Y_test, knn_prediction)
plt.plot(knn_fpr, knn_tpr, marker='.', label='KNN')
plt.xlabel('False Positive rate')
plt.ylabel('True Positive rate')
plt.legend()
plt.show()
print('KNN accuracy:', knn.score(X_test, Y_test))
```

# Random Forest accuracy on test data and ROC AUC

```python
random_forest_prediction = random_forest.predict_proba(X_test)
random_forest_prediction = random_forest_prediction[:, 1]
random_forest_auc = roc_auc_score(Y_test, random_forest_prediction)
print('Random Forest : ROC AUC=%.3f' % (random_forest_auc))
random_forest_fpr, random_forest_tpr, __noNeed = roc_curve(Y_test, random_forest_prediction)
plt.plot(random_forest_fpr, random_forest_tpr, marker='.', label='Random Forest')
plt.xlabel('False Positive rate')
plt.ylabel('True Positive rate')
plt.legend()
plt.show()
print('Random Forest accuracy:', random_forest.score(X_test, Y_test))
```
